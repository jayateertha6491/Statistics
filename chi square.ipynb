```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

chi square distribution is the distribution of a sum of the squares of k independent standard normal random variables. 

If Z1, ..., Zk are independent, standard normal random variables, then the sum of their squares,



Th resultant value have a distribution called chi square distribution denoted as 



```python
def generate_chisquare(dof):
    """
    The function creates normal distribution random variables and calculates chi square values
    from the normal distribution. Creates the graph for the given degrees of freedom for comparison
    """
    normal_data = {}
    for i in range(1,dof+1):
        normal_data[f"X{i}"] = np.random.randn(1000)
    
    ##normal_data = {1 : [normal_data],2:[normal_data]}....
    
    chi_square= {}
    for i in range(1,dof+1):
        normal_list = list(normal_data.keys())[:i]
        ab = [np.square(normal_data[i]) for i in normal_list]
        chi_square[f'chi_{i}'] = np.sum(ab, axis=0)
    
    plt.figure(figsize=(10,7))
    for k in chi_square.keys():
        sns.distplot(chi_square[k],hist = False,label=k)
    plt.xlim(0,5)
    plt.ylim(0,0.5)
    plt.legend()
    #return chi_square
```

```python
generate_chisquare(7)
```

```python
def generate_chisquare_lib(dof):
    chi_square = {}
    for i in range(1,dof+1):
        chi_square[f'chi{i}'] = np.random.chisquare(i,1000)
    plt.figure(figsize=(10,7))
    for k in chi_square.keys():
        sns.distplot(chi_square[k],hist = False,label=k)
    plt.xlim(0,5)
    #plt.ylim(0,0.5)
    plt.legend()
```

```python
generate_chisquare_lib(5)
```
